```python
vocab_size = meta["vocab_size"]
itos = meta["itos"]
stoi = meta["stoi"]

train_graph_idx = meta["train_data_graph_idx_list"]
val_graph_idx = meta["val_data_graph_idx_list"]
test_graph_idx = meta["test_data_graph_idx_list"]

emb_id_to_idx = meta["emb_graph_id_to_idx_dict"]
emb_idx_to_id = meta["emb_graph_idx_to_id_dict"]
```

```python
print("vocab_size:", vocab_size)
print("num train graphs:", len(train_graph_idx))
print("num val graphs:", len(val_graph_idx))
print("num test graphs:", len(test_graph_idx))

# roundtrip check
some_id = next(iter(emb_id_to_idx))
assert emb_idx_to_id[emb_id_to_idx[some_id]] == some_id
```
